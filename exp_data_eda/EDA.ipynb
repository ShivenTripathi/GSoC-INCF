```python
from scipy.sparse import load_npz
from hdnet import spikes
import numpy as np
from matplotlib import pyplot as plt
```

```python
ys = load_npz("MEA-37578-electricalStim-StimTimes-uint_N_61_1600_sparse.npz").toarray()
print(np.shape(ys))
x=spikes.Spikes(ys.T)
print(x)
print(np.shape(x._spikes))
```

```python
print(x.NOrderInteractions(1))
plt.matshow(x.NOrderInteractions(1), cmap='gray')
plt.show()
```

```python
store=x.get_frequencies()
```

```python
print(store.most_common(50))
```
